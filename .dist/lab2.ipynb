```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
```

```python
data=fetch_california_housing(as_frame=True)
california_data=data.frame
```

```python
correlation_matrix=california_data.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",fmt='.2f',linewidths=0.5)
plt.title("Correlation matrix of california_housing features")
plt.show()
```

```python
sns.pairplot(california_data,diag_kind='kde',plot_kws={'alpha':0.3})
plt.suptitle("pairplot for california features",y=1.02)
plt.show()-+
```
